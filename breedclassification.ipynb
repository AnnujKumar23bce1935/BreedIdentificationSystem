```python
import os
import shutil
import random
from pathlib import Path

# Paths
SOURCE_DIRS = [
    "/kaggle/input/indian-bovine-breeds",
    "/kaggle/input/indian-cattle-and-buffalo-breeds-dataset"
]

OUTPUT_DIR = "/kaggle/working/combined_livestock_dataset"

BREED_MAP = {
    "gir": "Gir",
    "gir_cow": "Gir",
    "sahiwal": "Sahiwal",
    "jersey": "Jersey",
    "murrah": "Murrah",
    "murrah_buffalo": "Murrah",
    "jaffarabadi": "Jaffarabadi"
}

SPLIT_RATIO = {"train": 0.7, "val": 0.15, "test": 0.15}

# Create folders
for split in SPLIT_RATIO:
    for breed in set(BREED_MAP.values()):
        Path(f"{OUTPUT_DIR}/{split}/{breed}").mkdir(parents=True, exist_ok=True)

def split_images(images):
    random.shuffle(images)
    n = len(images)
    return {
        "train": images[:int(0.7*n)],
        "val": images[int(0.7*n):int(0.85*n)],
        "test": images[int(0.85*n):]
    }

# Process datasets
for src in SOURCE_DIRS:
    for root, dirs, files in os.walk(src):
        folder = os.path.basename(root).lower()
        if folder in BREED_MAP:
            breed = BREED_MAP[folder]
            images = [os.path.join(root, f) for f in files if f.lower().endswith(('.jpg','.png','.jpeg'))]
            splits = split_images(images)

            for split, imgs in splits.items():
                for img in imgs:
                    shutil.copy(img, f"{OUTPUT_DIR}/{split}/{breed}")

print("✅ Dataset successfully combined and structured!")
```

```python
for split in ["train", "val", "test"]:
    print(f"\n{split.upper()}")
    for breed in os.listdir(f"{OUTPUT_DIR}/{split}"):
        count = len(os.listdir(f"{OUTPUT_DIR}/{split}/{breed}"))
        print(f"{breed}: {count} images")
```
